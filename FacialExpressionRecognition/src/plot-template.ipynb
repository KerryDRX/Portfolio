```python
# model comparison
val_acc1 = history1['val_model_acc']
val_acc2 = history2['val_model_acc']
val_acc3 = history3['val_model_acc']
val_acc4 = history4['val_model_acc']
val_acc0 = history0['val_model_acc']
val_acc5 = history5['val_model_acc']

epochs = range(40)

fig = plt.figure(figsize=(25, 10))
#plt.plot(epochs, val_acc0, label = "base")
plt.plot(epochs, val_acc1, label = "b+f")
plt.plot(epochs, val_acc2, label = "b+f+r10")
plt.plot(epochs, val_acc3, label = "b+f+r20")
plt.plot(epochs, val_acc4, label = "b+f+r30")
plt.plot(epochs, val_acc5, label = "b+f+r40")
plt.grid(linestyle = '--')
plt.title("Data Augmentation - Rotation",fontsize=16)
plt.xlabel('Epochs', fontsize=15)
plt.ylabel('Validation Accuracy', fontsize=15)
plt.ylim(0.83,0.865)


plt.legend()
plt.show()
```
